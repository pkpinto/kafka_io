````python
from multiprocessing import Process, Pipe

import BreadBasket_DMS_pb2
````

````python
class File2Conn(Process):
    '''class for batch processing of a csv data file'''

    def __init__(self, input, conn):
        Process.__init__(self)
        self.input = input
        self.conn = conn
        self._separator = ','


    def run(self):
        with open(self.input, 'r') as f:
            idx_map = {header: i for i, header in enumerate(f.readline().strip().split(self._separator))}
            for line in f:
                word = line.strip().split(self._separator)

                t = BreadBasket_DMS_pb2.Transaction()
                t.Date = word[idx_map['Date']]
                t.Time = word[idx_map['Time']]
                t.Transaction = int(word[idx_map['Transaction']])
                t.Item = word[idx_map['Item']]

                self.conn.send(t)
````

````python
class Conn2Stdout(Process):
    '''output class that prints what it gets to stdout'''

    def __init__(self, conn):
        Process.__init__(self)
        self.conn = conn


    def run(self):
        while True:
            print(self.conn.recv())
````

````python
conn_r, conn_w = Pipe(duplex=False)

input = File2Conn('data/BreadBasket_DMS.csv', conn_w)
input.start()
output = Conn2Stdout(conn_r)
output.start()
````
